```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
data = pd.read_csv('data_house.csv')
x = data['sqft_living'].to_numpy()
y = data['price'].to_numpy()
```

```python
m, c = np.random.random(), np.random.random()
L = 10e-8

epochs = 1000
for e in range(epochs):

    
    y_hat = m*x + c
    error = y - y_hat

    m += L*(error*x).mean()
    c += L*error.mean()

print(f'{m = :.2f}')
print(f'{c = :.2f}')
print(f'y = {m:.2f}x + {c:.2f}')
```

```python
yp = m*x+c
print(f'{yp = }')
```

```python
plt.scatter(x, y)
plt.plot(x, yp)
```
