```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
data = pd.read_csv('data.csv')
x = data['x'].to_numpy()
y = data['y'].to_numpy()
```

```python
def covar(x, y):
    return (x*y).mean() - x.mean()*y.mean()
```

```python
m = covar(x, y)/covar(x, x)
c = y.mean() - m*x.mean()

print(f'{m = :.2f}')
print(f'{c = :.2f}')
print(f'y = {m:.2f}x + {c:.2f}')
```

```python
yp = m*x+c
print(f'{yp = }')
```

```python
plt.scatter(x, y)
plt.plot(x, yp)
```

```python
m, c = 0, 0
L = 0.1

epochs = 100

for e in range(epochs):
    
    y_hat = m*x + c
    error = y - y_hat

    m += L*(error*x).mean()
    c += L*error.mean()

print(f'{m = :.2f}')
print(f'{c = :.2f}')
print(f'y = {m:.2f}x + {c:.2f}')
```

```python
yp = m*x+c
print(f'{yp = }')
```
